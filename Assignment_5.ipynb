#2. Build an HTML page with multiple paragraphs, each assigned a unique class name. Write CSS rules using class selectors to apply distinct styling to each paragraph. Follow the BEM naming convention and explain how you've named the classes.

```python
from IPython.display import HTML
html_code = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Q-2 & A-2</title>
    <link rel="stylesheet" href="styles2.css">
    <style>
/* Block: "text" */
.text {
    font-family: Arial, sans-serif;
    max-width: 600px;
    margin: 0 auto;
    padding: 20px;
}

/* Element: "text__paragraph" (base styling for all paragraphs) */
.text__paragraph {
    font-size: 16px;
    line-height: 1.5;
    margin-bottom: 20px;
}

/* Modifier: "text__paragraph--intro" (specific styling for the intro paragraph) */
.text__paragraph--intro {
    font-size: 20px;
    font-weight: bold;
    color: #333;
}

/* Modifier: "text__paragraph--highlight" (specific styling for the highlighted paragraph) */
.text__paragraph--highlight {
    background-color: yellow;
    padding: 10px;
    border: 2px solid orange;
}

/* Modifier: "text__paragraph--quote" (specific styling for the quote paragraph) */
.text__paragraph--quote {
    font-style: italic;
    border-left: 4px solid #ccc;
    padding-left: 10px;
    color: #555;
}
    </style>
</head>
<body>
    <!-- Block: "text" -->
    <div class="text">
        <!-- Element: "text__paragraph" -->
        <p class="text__paragraph text__paragraph--intro">
            This is the introductory paragraph. It has a larger font size and a bold weight.
        </p>
        <!-- Element: "text__paragraph" with modifier "text__paragraph--highlight" -->
        <p class="text__paragraph text__paragraph--highlight">
            This paragraph is highlighted with a yellow background and a border.
        </p>
        <!-- Element: "text__paragraph" with modifier "text__paragraph--quote" -->
        <p class="text__paragraph text__paragraph--quote">
            This paragraph is styled as a quote with italic text and a left border.
        </p>
    </div>
</body>
</html>
 """
HTML(html_code)
```

Answer: Explanation of BEM Naming Convention
Block:

A standalone component or section of the page (e.g., .text).

Represents a higher-level container for elements.

Element:

A part of the block that performs a specific function (e.g., .text__paragraph).

Elements are named using the block name followed by __ and the element name.

Modifier:

A variation of a block or element (e.g., .text__paragraph--intro).

Modifiers are named using the block or element name followed by -- and the modifier name.

How Classes Are Named in This Example
Block: .text

Represents the container for all paragraphs.

Element: .text__paragraph

Represents the base styling for all paragraphs within the .text block.

Modifiers:

.text__paragraph--intro: Applies specific styling for the introductory paragraph.

.text__paragraph--highlight: Applies specific styling for the highlighted paragraph.

.text__paragraph--quote: Applies specific styling for the quote paragraph.

#3. Develop an HTML form with various input elements. Use CSS to style the form, including setting background colors for input fields. Create a custom color palette for the form elements, and demonstrate how to apply opacity to one of the form sections.

```python
from IPython.display import HTML
html_code = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Q-3 & A-3</title>
    <link rel="stylesheet" href="styles3.css">
    <style>
  /* Custom Color Palette */
:root {
    --primary-color: #4a90e2; /* Blue */
    --secondary-color: #50e3c2; /* Teal */
    --background-color: #f5f5f5; /* Light Gray */
    --text-color: #333; /* Dark Gray */
    --accent-color: #ff6f61; /* Coral */
}

/* Form Container */
.form-container {
    max-width: 600px;
    margin: 0 auto;
    padding: 20px;
    background-color: var(--background-color);
    border-radius: 10px;
    box-shadow: 0 4px 8px rgba(0, 0, 0, 0.1);
}

/* Form Section */
.form-section {
    margin-bottom: 20px;
    padding: 15px;
    background-color: white;
    border-radius: 8px;
    border: 1px solid #ddd;
}

/* Form Section with Opacity */
.form-section--opacity {
    opacity: 0.8; /* Apply opacity */
    background-color: var(--secondary-color);
    color: white;
}

/* Labels */
label {
    display: block;
    margin-bottom: 5px;
    font-weight: bold;
    color: var(--text-color);
}

/* Input Fields */
input[type="text"],
input[type="email"],
input[type="tel"],
input[type="date"],
input[type="color"],
textarea {
    width: 100%;
    padding: 10px;
    margin-bottom: 15px;
    border: 1px solid #ddd;
    border-radius: 5px;
    font-size: 16px;
    background-color: var(--background-color);
    color: var(--text-color);
}

/* Focus State for Input Fields */
input:focus,
textarea:focus {
    border-color: var(--primary-color);
    outline: none;
    box-shadow: 0 0 5px rgba(74, 144, 226, 0.5);
}

/* Submit Button */
button {
    width: 100%;
    padding: 10px;
    background-color: var(--primary-color);
    color: white;
    border: none;
    border-radius: 5px;
    font-size: 16px;
    cursor: pointer;
}

button:hover {
    background-color: var(--accent-color);
}
    </style>
</head>
<body>
    <div class="form-container">
        <h1>Contact Us</h1>
        <form>
            <!-- Section 1: Basic Information -->
            <div class="form-section">
                <h2>Basic Information</h2>
                <label for="name">Name:</label>
                <input type="text" id="name" name="name" placeholder="Enter your name" required>

                <label for="email">Email:</label>
                <input type="email" id="email" name="email" placeholder="Enter your email" required>
            </div>

            <!-- Section 2: Additional Information (with opacity) -->
            <div class="form-section form-section--opacity">
                <h2>Additional Information</h2>
                <label for="phone">Phone:</label>
                <input type="tel" id="phone" name="phone" placeholder="Enter your phone number">

                <label for="dob">Date of Birth:</label>
                <input type="date" id="dob" name="dob">

                <label for="color">Favorite Color:</label>
                <input type="color" id="color" name="color" value="#ff0000">
            </div>

            <!-- Section 3: Message -->
            <div class="form-section">
                <h2>Your Message</h2>
                <label for="message">Message:</label>
                <textarea id="message" name="message" placeholder="Enter your message" rows="5" required></textarea>
            </div>

            <!-- Submit Button -->
            <button type="submit">Submit</button>
        </form>
    </div>
</body>
</html>
 """
HTML(html_code)
```
